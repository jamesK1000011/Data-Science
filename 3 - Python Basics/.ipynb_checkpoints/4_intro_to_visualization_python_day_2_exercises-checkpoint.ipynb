#### Question 1
##### Let's create a simple scatterplot using plotly's native syntax. 
##### Create `trace` using `go.Scatter()` and variables `hardship_index` as `x` and `percent_house_below_poverty` as `y`.
##### Label `title`, `xaxis` and `yaxis` in `layout`.
##### Create a dictionary with both parts and save as `fig`. 
##### Use `iplot()` to plot the graph.

```python
# Create a trace.
trace = go.Scatter(
    x = chicago_census ['hardship_index'],
    y = chicago_census['percent_house_below_poverty'],
    mode = 'markers'
)

data = [trace]

layout = {
    'title':'Simple Scatterplot',
    'xaxis': { 'title' : 'hardship index', 'zeroline': False },
    'yaxis': { 'title' : '% below poverty', 'zeroline' : False}
}

fig = {'data':data, 'layout':layout}

# Plot and embed in iPython notebook!
iplot(fig, filename='cc_basic-scatter')
```

#### Question 3
##### Let's prepare data for our layered bar chart.
##### Group `chicago_census` data by `per_capita_income` variable. Save as `ex_grouped`.
##### Then group and summarize the numeric variables  `percent_house_crowded`, 
##### `percent_house_below_poverty`,`percent_16_unemployed`, `percent_25_without_diploma`, `percent_dependent`, 
###### `hardship_index` by `per_capita_income` using their means. Save as `ex_grouped_mean` and print.
##### Reset its index and print the result.



```python
ex_grouped = chicago_census.groupby('per_capita_income')
# Compute mean on the listed variables using the grouped data.
ex_grouped_mean = ex_grouped.mean()[['percent_house_crowded','percent_house_below_poverty','percent_16_unemployed','percent_25_without_diploma', 'percent_dependent',
'hardship_index']]
ex_grouped_mean
# Reset index of the dataset.
ex_grouped_mean = ex_grouped_mean.reset_index()
ex_grouped_mean

```

#### Question 4
##### Use the `pd.melt()` function and convert `ex_grouped_mean` to long format. Save as `ex_grouped_mean_long` and print the result.
##### Filter `per_capita_income` as 1 and then create two columns in the dataframe: `metrics` and `means`.
##### Save as `ex_true_means` and print.
##### Do the same filter for `per_capita_income` as 0 and save as `ex_false_means`.

```python
ex_grouped_mean_long = pd.melt(ex_grouped_mean,       #<- wide dataset
id_vars = ['per_capita_income'],     #<- identifying variable
var_name = 'metric',      #<- contains col names of wide data
value_name = 'mean')      #<- contains values from above columns
#ex_grouped_mean_long

ex_true_means = ex_grouped_mean_long.query('per_capita_income == 1')[['metric', 'mean']]
print(ex_true_means)
# Let's get the `Target` = `False` mean data.
ex_false_means = ex_grouped_mean_long.query('per_capita_income == 0')[['metric', 'mean']]
print(ex_false_means)
```

##### Create a simple cholopleth map of `pop` and set `text` as `country`
##### Set `locations` as `iso_alpha`.

```python
gdp = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminder_with_codes.csv')
```

```python
gdp.head()
```

```python
choropleth_trace = go.Choropleth(
    locations = gdp['iso_alpha'],
    z = gdp['pop'],
    text = gdp['country'],
)
data = [choropleth_trace]
iplot(data, filename = 'pop-map')
```

#### Answer:

#### Question 2
##### Customize the map by adding `colorscale` as shown below:
colorscale = [
        [0, "rgb(171, 81, 22)"],
        [0.5,"rgb(70, 99, 78)"],
        [1, "rgb(88,199,118)"]
        
##### You can also try out your own colorscale and set the number of steps between the color change as you desire.
##### Add colorbar title as `Population` and plot the map.

```python
choropleth_trace = go.Choropleth(
    locations = gdp['iso_alpha'],
    z = gdp['pop'],
    text = gdp['country'],
    colorscale = [
        [0, "hotpink"],
        [0.5, "purple"],
        [1, "white"]
    ],
    autocolorscale = False,
    reversescale = True,
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'rgb(180,180,180)',width = 0.5
        )
    ),
    colorbar = go.choropleth.ColorBar(
        title = '#Population'
    ),
)

data = [choropleth_trace]

layout = go.Layout(
    title = go.layout.Title(
        text = 'Global Population'
    )
)
```

```python
fig = go.Figure(data = data, layout = layout)
iplot(fig, filename = 'pop-map')
```

#### Question 3
##### Change the projection of the above map to type `hammer` in attribute `geo` of `layout`.

```python
layout = go.Layout(
    title = go.layout.Title(
        text = 'Global Population'
    ),
    geo = go.layout.Geo(
        scope = 'world',
        projection = go.layout.geo.Projection(type = 'hammer')
    )
)

fig = go.Figure(data = data, layout = layout)
iplot(fig, filename = 'world-map')
```

#### Question 4
##### Now let's create a map of USA. Read in `state_data.csv` and name as `state_data`.
##### Set the colorscale `scl` as shown in class. 
##### Create a choropleth map using variable `Frost`.
##### Set locations as `code` and `text` as `state`.

```python
state_data = pd.read_csv('state_data.csv')
state_data.head()
scl = [
    [0.0, 'rgb(242,240,247)'],
    [0.2, 'rgb(218,218,235)'],
    [0.4, 'rgb(188,189,220)'],
    [0.6, 'rgb(158,154,200)'],
    [0.8, 'rgb(117,107,177)'],
    [1.0, 'rgb(84,39,143)']
]
```

```python
trace = go.Choropleth(
    colorscale = scl,
    autocolorscale = False,
    reversescale = False,
    locations = state_data['code'],
    z = state_data['Frost'],
    locationmode = 'USA-states',
    text = state_data['State'],
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'white',
            width = 2
        )
    ),
    colorbar = go.choropleth.ColorBar(
        title = "Frost"
    )
)

data = [trace]

layout = go.Layout(
    title = go.layout.Title(
    text = 'Winter is Coming'
),
    geo = go.layout.Geo(
        scope = 'usa',
        projection = go.layout.geo.Projection(type = 'albers usa'),
        showlakes = True,
        lakecolor = 'White'),
)

fig = go.Figure(data = data, layout = layout)
iplot(fig, filename = 'usa-map')

```
