## Exercise 3

#### Question 1
##### We will scrape the entire table and store to a dataframe in this section.
##### Save the rows in the table as `rows` by selecting the tag `tr` of `table[1]`.
##### We need to access `table[1]` because `table` is a list, as we created it with `.select()` in Question 3 of Exercise 3.
##### Find the length of the `rows`


```python
rows = table[1].select('tr')
print(rows)
len(rows)
```

#### Question 2
##### Create 5 empty lists with name `rank`, `country_name`, `region`, `population`, `change`.
##### We will scrape the table now. Follow the instructions carefully step by step.
##### The for loop should extend from 3rd row (index 2) till the end (len(rows)).
##### We will access each element of interest in the row by selecting the tag `td` with the index of the element of interest
##### First, call each row by iterating in the for loop and select `td` tag.
##### We want rank, country, region, recent population and change which are stored in index 0, 1, 2, 5, 6 respectively.
##### Access each td tag using the index element and fetch the text of each element and append to respective list.


```python
rank = []
country_name = []
region = []
population = []
change = []
```

```python

for row in range(2,len(rows)):
    # Extract the desired elements from every row.
    rk = rows[row].select('td')[0].text
    cn = rows[row].select('td')[1].text.strip()
    rn = rows[row].select('td')[2].text.strip()
    pop = rows[row].select('td')[5].text.strip()
    chg = rows[row].select('td')[6].text.strip()
    # Append extracted elements to lists.
    rank.append(rk)
    country_name.append(cn)
    region.append(rn)
    population.append(pop)
    change.append(chg)
```

#### Question 3
##### Print first 10 elements of each list.


```python
print(rank[:10])
print(country_name[:10])
print(region[:10])
print(population[:10])
print(change[:10])
```

#### Question 4
##### We will clean the lists by removing special characters in lists rank, country_name, population and change.
##### Slice all special characters of each elements of each list.
##### Use code: [x[:index] for x in list_name].


```python
rank = [x[:-1] for x in rank]
```

```python
rank
```

#### Question 5
##### Import pandas.
##### Combine all the list as dataframe named 'df_ex' and display the head of it.


```python

print()
df_ex = pd.DataFrame(list(zip(rank,
                              country_name,
                              region,
                              population,
                              change)),
                    columns =['rank',
                              'country_name',
                              'region',
                              'population',
                              'change'])
df_ex['population'] = df_ex['population'].apply(lambda x: x.replace(',', ''))
df_ex.head()
```

#### Question 6

##### As you can see, the first entry in the column `Country` still contains some unnecessary characters. Replace this entry with `China`.

##### Display the updated dataframe.


```python
df_ex = df_ex.replace({'China[a]': 'China'})
df_ex.head()
```
