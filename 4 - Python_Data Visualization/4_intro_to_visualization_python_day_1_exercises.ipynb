#### Question 4
##### Drop columns `community_number` and `community_area` from the dataframe.

#### Question 5
##### Create a variable `mean_per_capita_income` which contains the mean of `per_capita_income` from `chicago_census`.
##### Convert `per_capita_income` variable to a binary variable where `per_capita_income` is set to 0 if per_capita_income of the row is less than `mean per_capita_income`, otherwise set `per_capita_income` to 1.
###### Make a new duplicate dataframe called `ex_viz`.

```python
chicago_census['mean_per_capita_income'] = chicago_census.mean()['per_capita_income']
chicago_census.head()
```

```python
chicago_census['per_capita_income'] = np.where(chicago_census['per_capita_income'] > chicago_census['mean_per_capita_income'], 1, 0)
chicago_census.head()
```

```python
ex_viz = chicago_census
ex_viz.head()
```

#### Question 6

##### Group `ex_viz` data by the `per_capita_income` variable. Save as `ex_grouped`.
##### Then group and summarize the numeric variables `community_number`, `percent_house_crowded`, 
##### `percent_house_below_poverty`,`percent_16_unemployed`, `percent_25_without_diploma`, `percent_dependent`, 
###### `hardship_index` by `per_capita_income` using their means. Save as `ex_grouped_mean` and print.
##### Reset its index and print the result.

```python
ex_grouped = ex_viz.groupby('per_capita_income')
ex_grouped_mean = ex_grouped.mean()[['percent_house_crowded',
                                     'percent_house_below_poverty',
                                     'percent_16_unemployed',
                                     'percent_25_without_diploma',
                                     'percent_dependent',
                                     'hardship_index']]
ex_grouped_mean.head()
```

```python
ex_grouped_mean = ex_grouped_mean.reset_index()
ex_grouped_mean.head()
```

#### Question 7

##### Notice the format of `ex_grouped_mean`. We wish to convert it from wide to long format.
##### Use the `pd.melt()` function and convert it to long format. Save as `ex_grouped_mean_long` and print the result.

```python
ex_grouped_mean_long = pd.melt(ex_grouped_mean, 
                                  id_vars = ['per_capita_income'], #<- identifying variable
                                  var_name = 'metric', #<- contains col names of wide data 
                                  value_name = 'mean') #<- contains values from above columns

ex_grouped_mean_long
```

#### Question 8
##### Now use the `pd.pivot()` function to convert `ex_grouped_mean_long` to wide format.
##### Save as `ex_grouped_mean_wide` and print.

```python
ex_grouped_mean_wide = ex_grouped_mean_long.pivot(
                    index = 'per_capita_income', #<- identifying variable
                    columns = 'metric', #<- col names of wide data
                    values = 'mean') #<- values from above columns
ex_grouped_mean_wide
```

#### Question 3
##### Let's create a multiple layered scatterplot using the `plt.subplots()` function.
##### Define the `key` and `value` for the color dictionary `ex_color_dict`. 
##### Set the `per_capita_income` as `Below Mean Income` and `Above Mean Income` for key value `0` and `1` respectively.
##### View the plot.
##### Hint: It's always better to clear the plotting area before rendering a new plot by using `plt.clf()`.

```python
ex_viz.head()
```

```python
plt.clf()
fig, axes = plt.subplots() 
for key, value in ex_color_dict.items():
    percent_16_unemployed = ex_viz.query('per_capita_income=='+str(key))['percent_16_unemployed']
    hardship_index = ex_viz.query('per_capita_income=='+str(key))['hardship_index']
    
    if key == 0:
        Flag = "Below Mean Income" 
    else:
        Flag = "Above Mean Income" 
        
    axes.scatter(percent_16_unemployed,
                 hardship_index,
                 c = value,
                 label = Flag,
                 alpha = 0.3) 
             
axes.legend() #<- add a legend to get labels
plt.show()
```
